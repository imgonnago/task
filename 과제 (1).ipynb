```python
train_df = pd.read_csv("/Users/joyongjae/dataset_fruit/train.csv")
test_df = pd.read_csv("/Users/joyongjae/dataset_fruit/test.csv")
train_df.head()
```

```python
#수관폭평균, 새순은 전부 사용, 엽록소는 10~11월 데이터만 사용 / mae value: mae value: 31.516123680241332 / nmae value: 0.07656720410406744
x_train = train_df.drop(['ID','수고(m)','착과량(int)','수관폭1(min)','수관폭2(max)'], axis=1)
x_test = test_df.drop(['ID','수고(m)','착과량(int)','수관폭1(min)','수관폭2(max)'], axis=1)
print(x_train.columns[120])
print(x_train.columns[62])
x_train = x_train.drop(x_train.iloc[:, 90:120], axis=1)
x_test = x_test.drop(x_test.iloc[:, 90:120], axis=1)

print(x_train)
y_train = train_df['착과량(int)']
y_test = test_df['착과량(int)']
```

```python
model = RandomForestRegressor(random_state=42)

model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
mae = mean_absolute_error(y_test, pred)
nmae = mae / np.mean(np.abs(y_test))
print(f" mae value: {mae} / nmae value: {nmae}")
```
